```python
'''



시계열 데이터란?

시계열 데이터는 시간에 따라 순서대로 수집된 데이터입니다.
즉, 데이터를 기록한 시점(시간)과 그에 해당하는 값이 포함된 데이터 집합을 의미합니다.
시계열 데이터는 일정한 시간 간격(예: 일별, 월별, 연도별 등)으로 측정되며, 시간의 흐름에 따른 패턴을 분석하는 데 사용됩니다.

예시:

주식 가격: 주식의 일별 종가를 기록한 데이터는 시계열 데이터입니다.
예를 들어, 삼성전자 주식의 매일 종가를 1년 동안 기록한 데이터가 이에 해당합니다.

기온 데이터: 특정 지역의 일일 기온을 기록한 데이터도 시계열 데이터입니다.
예를 들어, 서울의 하루 평균 기온을 2023년 1월부터 12월까지 기록한 데이터입니다.

판매량: 특정 제품의 월별 판매량을 기록한 데이터도 시계열 데이터입니다.
예를 들어, 2023년 동안의 월별 스마트폰 판매량 데이터입니다.



시계열 분석의 필요성

시계열 분석은 시간에 따른 데이터의 변화를 이해하고 예측하기 위한 기법입니다.
이를 통해 데이터의 패턴을 파악하고, 미래의 값을 예측하거나, 특정 사건의 영향을 분석할 수 있습니다.
시계열 분석은 비즈니스 의사결정, 재무 계획, 정책 수립 등 다양한 분야에서 중요한 역할을 합니다.

예시:

주식 시장 분석: 투자자는 주식의 과거 가격 변동을 분석하여 미래의 주가를 예측합니다. 이는 투자 전략을 세우는 데 도움이 됩니다.

기후 변화 예측: 기후학자는 장기적인 기온 변화를 분석하여 기후 변화를 예측하고, 이로 인해 발생할 수 있는 환경적 영향을 평가합니다.

판매 예측: 기업은 과거의 판매 데이터를 분석하여 향후 판매량을 예측합니다. 이를 통해 재고 관리, 생산 계획 등을 최적화할 수 있습니다.



시계열 데이터의 특성

추세 (Trend): 데이터가 장기적으로 증가하거나 감소하는 패턴을 나타냅니다.
예를 들어, 기술 발전으로 인해 스마트폰 판매량이 지속적으로 증가하는 경우가 해당됩니다.

계절성 (Seasonality): 데이터가 특정 주기(계절, 월, 요일 등)에 따라 반복되는 패턴을 보여줍니다.
예를 들어, 여름에 에어컨 판매량이 증가하는 패턴이 이에 해당합니다.

사이클 (Cycle): 데이터가 장기적인 경제적 또는 비즈니스 사이클에 따라 변동하는 패턴입니다.
예를 들어, 경기 침체와 호황 주기에 따라 기업의 매출이 변화하는 경우입니다.

불규칙성 (Irregularity): 예측할 수 없는 랜덤한 변동을 나타냅니다.
예를 들어, 갑작스런 자연재해로 인해 발생하는 매출 감소 등이 이에 해당합니다.


'''
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# 데이터 생성
np.random.seed(0)  # 랜덤 데이터의 재현성을 위해 시드 설정

# 월별 날짜 생성 (2022년 1월부터 2024년 12월까지)
months = pd.date_range(start='2022-01-01', end='2024-12-01', freq='MS')
n_months = len(months)  # 총 월 수 계산

# 추세 생성: 50에서 150까지 선형적으로 증가
trend = np.linspace(50, 150, n_months)

# 계절성 생성: 12개월 주기로 반복되는 사인 함수
seasonality = 15 * np.sin(np.linspace(0, 2 * np.pi, n_months))

# 순환성 생성: 약 24개월 주기의 순환 패턴 추가
cycle = 10 * np.sin(np.linspace(0, 2 * np.pi * 2, n_months))

# 불규칙성 생성: 평균 0, 표준편차 7의 정규 분포에서 랜덤 노이즈 추가
noise = np.random.normal(0, 7, n_months)

# 월별 판매량 데이터 생성: 추세 + 계절성 + 순환성 + 노이즈
sales = trend + seasonality + cycle + noise

# 데이터프레임 생성
data = pd.DataFrame({'Date': months, 'Sales': sales})
data.set_index('Date', inplace=True)

# 원본 데이터 시각화
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')
plt.title('Original Sales Data')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
'''


1. 추세 (Trend)

추세는 데이터의 장기적인 방향성을 나타냅니다.
시간이 지남에 따라 데이터가 증가하거나 감소하는 경향을 포착합니다.

단계:

이동 평균 계산: 데이터의 이동 평균을 계산하여 평활화합니다. 이동 평균은 특정 기간 동안의 데이터 값을 평균하여 데이터를 평활화하는 방법입니다.

회귀 분석: 데이터의 추세를 모델링하기 위해 선형 회귀를 사용할 수도 있습니다. 이 경우, 시간에 따른 데이터의 선형적 증가 또는 감소를 추정합니다.

결과: 추세 성분은 장기적인 경향을 나타내며, 일반적으로 데이터의 큰 방향성을 보여줍니다.


2. 계절성 (Seasonality)

계절성은 일정 주기마다 반복되는 패턴을 나타냅니다. 데이터에서 계절성은 특정 기간(예: 월, 분기)마다 반복되는 패턴입니다.

단계:

주기 설정: 계절성을 분석하기 위해 데이터의 주기를 설정합니다. 예를 들어, 월별 데이터에서는 12개월 주기를 설정할 수 있습니다.

평활화: 추세를 제거한 후 데이터를 평활화하여 계절성을 추출합니다. 계절성은 반복적인 패턴을 강조합니다.

결과: 계절성 성분은 일정 주기로 반복되는 패턴을 나타내며, 계절적 변화나 주기적인 패턴을 보여줍니다.


3. 순환성 (Cycle)

순환성은 계절성보다 긴 주기를 가진 반복적인 패턴을 나타냅니다. 이는 경제적 주기나 기타 장기적 변동을 반영할 수 있습니다.

추세와 계절성 제거: 데이터를 평활화하여 추세와 계절성을 제거한 후, 남은 부분에서 순환성을 추출합니다.

순환성 분석: 잔여 데이터를 분석하여 장기적인 주기를 가진 반복적인 패턴을 식별합니다. 예를 들어, 24개월 주기의 순환성 패턴을 식별할 수 있습니다.

결과: 순환성 성분은 장기적인 반복 패턴을 나타내며, 경제적 변화나 다른 장기적 요인을 반영할 수 있습니다.


4. 불규칙성 (Residuals)

불규칙성은 시계열 데이터에서 추세, 계절성, 순환성을 제외한 나머지 변동을 나타냅니다. 이는 예측할 수 없는 변동성이나 노이즈를 포함합니다.

추세, 계절성, 순환성 제거: 원본 데이터에서 추세, 계절성, 순환성을 제거하여 남은 데이터를 불규칙성으로 계산합니다.

잔여 데이터 분석: 잔여 데이터를 분석하여 불규칙성을 평가합니다. 이는 데이터의 예측할 수 없는 변동성이나 노이즈를 포함합니다.

결과: 불규칙성 성분은 데이터를 예측할 수 없는 부분을 나타내며, 일반적으로 노이즈나 비정상적인 변동을 보여줍니다.


'''


from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# 시계열 데이터 분해
decomposition = seasonal_decompose(data['Sales'], model='additive', period=12)  # 데이터의 주기는 12개월

# 추세, 계절성, 불규칙성 추출
trend = decomposition.trend  # 추세 컴포넌트
seasonality = decomposition.seasonal  # 계절성 컴포넌트
residual = decomposition.resid  # 불규칙성(잔차) 컴포넌트

# 순환성 추출 (순환성은 수동으로 추출해야 함)
# 순환성의 주기가 약 24개월이라고 가정하여 필터링
cycle_period = 24
cycle_component = (data['Sales'] - trend - seasonality)  # 추세와 계절성을 제외한 부분

# 데이터 시각화
plt.figure(figsize=(14, 12))

# 원본 데이터
plt.subplot(5, 1, 1)
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')
plt.title('Original Sales Data')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# 추세
plt.subplot(5, 1, 2)
plt.plot(data.index, trend, label='Trend', color='r')
plt.title('Trend Component')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# 계절성
plt.subplot(5, 1, 3)
plt.plot(data.index, seasonality, label='Seasonality', color='g')
plt.title('Seasonality Component')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# 순환성
plt.subplot(5, 1, 4)
plt.plot(data.index, cycle_component, label='Cycle Component', color='c')
plt.title('Cycle Component')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# 불규칙성
plt.subplot(5, 1, 5)
plt.plot(data.index, residual, label='Residual', color='m')
plt.title('Residual Component')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

plt.tight_layout()  # 서브플롯 간의 간격 조정
plt.show()
```

```python
'''

4. 시계열 데이터의 전처리
4.1. 결측치 처리
4.2. 이상치 탐지 및 처리
4.3. 데이터 변환 (로그 변환, 차분 등)

'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 데이터 생성
np.random.seed(0)  # 랜덤 데이터의 재현성을 위해 시드 설정
dates = pd.date_range(start='2022-01-01', end='2024-12-01', freq='MS')
n = len(dates)
data = pd.DataFrame({
    'Date': dates,
    'Sales': 100 + np.linspace(0, 50, n) + 20 * np.sin(np.linspace(0, 2 * np.pi, n)) + np.random.normal(0, 10, n)
})
data.set_index('Date', inplace=True)

# 인위적인 결측치 및 이상치 추가
data.loc['2023-06-01', 'Sales'] = np.nan  # 결측치
data.loc['2023-07-01', 'Sales'] = 300  # 이상치

# 데이터 시각화
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')
plt.title('Original Sales Data with Missing Values and Outliers')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
'''


결측치 처리 방법 (여러가지중 하나)

선형 보간법 (Linear Interpolation):

결측치가 있는 시점과 인접한 시점의 데이터 값을 연결하여 직선 방정식을 구성합니다.
결측치는 이 직선 방정식을 사용하여 추정됩니다.
interpolate() 함수를 사용하여 결측치를 선형 보간법으로 채울 수 있습니다.

평균값 대체 (Mean Imputation):

결측치를 전체 데이터의 평균값으로 대체하는 방법입니다.
데이터의 평균값을 계산하고, 결측치가 있는 곳에 이 값을 대입합니다.
fillna() 함수를 사용하여 결측치를 평균값으로 대체합니다.


'''

# 결측치 처리

# 결측치 처리 - 선형 보간법
data['Sales_interpolated'] = data['Sales'].interpolate(method='linear')

# 결측치 처리 - 평균값으로 대체
mean_value = data['Sales'].mean()
data['Sales_filled'] = data['Sales'].fillna(mean_value)

# 시각화
plt.figure(figsize=(12, 6))

# 원본 데이터
plt.subplot(3, 1, 1)
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')
plt.title('Original Sales Data with Missing Values')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# 선형 보간법으로 결측치 처리
plt.subplot(3, 1, 2)
plt.plot(data.index, data['Sales_interpolated'], label='Interpolated Sales Data', color='g')
plt.title('Sales Data after Linear Interpolation')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# 평균값으로 결측치 대체
plt.subplot(3, 1, 3)
plt.plot(data.index, data['Sales_filled'], label='Filled Sales Data', color='r')
plt.title('Sales Data with Missing Values Filled with Mean')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
'''

IQR (Interquartile Range):

데이터의 중앙값을 기준으로 이상치를 탐지합니다.
IQR은 데이터의 1사분위수(Q1)와 3사분위수(Q3)의 차이입니다.
IQR 범위를 벗어난 값들을 이상치로 간주합니다.

Z-score:

데이터 값이 평균으로부터 얼마나 떨어져 있는지를 표준편차 단위로 측정합니다.
일반적으로 Z-score가 3보다 크면 이상치로 간주합니다.


'''
# 이상치 탐지


from scipy import stats

# IQR을 이용한 이상치 탐지
Q1 = data['Sales'].quantile(0.25)
Q3 = data['Sales'].quantile(0.75)
IQR = Q3 - Q1
outlier_mask = (data['Sales'] < (Q1 - 1.5 * IQR)) | (data['Sales'] > (Q3 + 1.5 * IQR))

# Z-score를 이용한 이상치 탐지
z_scores = np.abs(stats.zscore(data['Sales'].dropna()))
outliers_z = (z_scores > 3)

# 이상치 처리 - IQR 기준으로 이상치 제거
data['Sales_no_outliers'] = data['Sales'].where(~outlier_mask, np.nan)
data['Sales_no_outliers'] = data['Sales_no_outliers'].interpolate(method='linear')

# 이상치 처리 - Z-score 기준으로 이상치 제거
data['Sales_no_outliers_z'] = data['Sales'].where(~outliers_z, np.nan)
data['Sales_no_outliers_z'] = data['Sales_no_outliers_z'].interpolate(method='linear')

# 시각화
plt.figure(figsize=(12, 8))

# 원본 데이터
plt.subplot(4, 1, 1)
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')
plt.title('Original Sales Data with Outliers')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# IQR 기준으로 이상치 처리
plt.subplot(4, 1, 2)
plt.plot(data.index, data['Sales_no_outliers'], label='Sales Data after IQR Outlier Removal', color='g')
plt.title('Sales Data after IQR Outlier Removal')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# Z-score 기준으로 이상치 처리
plt.subplot(4, 1, 3)
plt.plot(data.index, data['Sales_no_outliers_z'], label='Sales Data after Z-score Outlier Removal', color='r')
plt.title('Sales Data after Z-score Outlier Removal')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
'''


데이터 변환
데이터 변환은 데이터를 더 분석하기 쉬운 형태로 변환하거나, 데이터의 분포를 조정하여 분석을 용이하게 합니다. 주요 방법은 다음과 같습니다:

로그 변환 (Log Transformation):

데이터의 분포가 비정규적이거나 매우 큰 변동성을 가지는 경우, 로그 변환을 통해 분포를 더 정규에 가깝게 만듭니다.
로그 변환은 큰 값의 영향을 줄이고, 데이터를 스케일링하여 더 안정적인 분석을 가능하게 합니다.

차분 (Differencing):

시계열 데이터에서 추세나 계절성을 제거하고, 데이터의 차분을 통해 안정적인 시계열을 만듭니다.
첫 번째 차분은 현재 값과 이전 값의 차이를 계산하여 시계열을 더 안정적으로 만듭니다.


'''
from scipy import stats

# IQR을 이용한 이상치 탐지
Q1 = data['Sales'].quantile(0.25)
Q3 = data['Sales'].quantile(0.75)
IQR = Q3 - Q1
outlier_mask = (data['Sales'] < (Q1 - 1.5 * IQR)) | (data['Sales'] > (Q3 + 1.5 * IQR))

# Z-score를 이용한 이상치 탐지
z_scores = np.abs(stats.zscore(data['Sales'].dropna()))
outliers_z = (z_scores > 3)

# 이상치 처리 - IQR 기준으로 이상치 제거
data['Sales_no_outliers'] = data['Sales'].where(~outlier_mask, np.nan)
data['Sales_no_outliers'] = data['Sales_no_outliers'].interpolate(method='linear')

# 이상치 처리 - Z-score 기준으로 이상치 제거
data['Sales_no_outliers_z'] = data['Sales'].where(~outliers_z, np.nan)
data['Sales_no_outliers_z'] = data['Sales_no_outliers_z'].interpolate(method='linear')

# 시각화
plt.figure(figsize=(12, 8))

# 원본 데이터
plt.subplot(4, 1, 1)
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')
plt.title('Original Sales Data with Outliers')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# IQR 기준으로 이상치 처리
plt.subplot(4, 1, 2)
plt.plot(data.index, data['Sales_no_outliers'], label='Sales Data after IQR Outlier Removal', color='g')
plt.title('Sales Data after IQR Outlier Removal')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

# Z-score 기준으로 이상치 처리
plt.subplot(4, 1, 3)
plt.plot(data.index, data['Sales_no_outliers_z'], label='Sales Data after Z-score Outlier Removal', color='r')
plt.title('Sales Data after Z-score Outlier Removal')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
'''

자기상관과 부분자기상관

자기상관 (ACF):

자기상관 함수(ACF)는 시간 지연에 따라 시계열 데이터의 값들이 얼마나 상관되어 있는지를 측정합니다.
ACF는 시계열의 데이터와 지연된 버전 사이의 상관관계를 나타냅니다.
부분자기상관 (PACF):

부분자기상관 함수(PACF)는 특정 지연 시점에서의 상관관계와 다른 모든 중간 지연 시점의 영향을 제거한 상관관계를 측정합니다.
PACF는 자기회귀 모델의 적합에 유용합니다.
원리 및 단계
자기상관 (ACF):

데이터와 지연된 데이터 간의 상관계수를 계산합니다.
지연 시점의 상관계수를 시각화하여 시계열의 패턴을 이해합니다.
부분자기상관 (PACF):

자기상관에서 중간 지연의 영향을 제거한 순수한 상관관계를 계산합니다.
부분자기상관을 시각화하여 자기회귀 모델의 적합에 도움을 줍니다.

'''


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 예시 데이터 생성
np.random.seed(0)
dates = pd.date_range(start='2022-01-01', end='2024-12-01', freq='MS')
n = len(dates)
data = pd.DataFrame({
    'Date': dates,
    'Sales': 100 + np.linspace(0, 50, n) + 20 * np.sin(np.linspace(0, 2 * np.pi, n)) + np.random.normal(0, 10, n)
})
data.set_index('Date', inplace=True)

# 결측치가 있는지 확인하고 처리
data['Sales'] = data['Sales'].fillna(method='ffill')  # 결측치가 있으면 선형 보간 또는 이전 값으로 대체

# 자기상관 및 부분자기상관 분석
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# 자기상관 (ACF) - 시계열의 자기상관 함수 플롯
plot_acf(data['Sales'], lags=min(40, len(data) // 2), ax=axes[0])  # 40 시차까지 자기상관 플롯
axes[0].set_title('Autocorrelation Function (ACF)')  # 그래프 제목 설정

# 부분자기상관 (PACF) - 시계열의 부분자기상관 함수 플롯
plot_pacf(data['Sales'], lags=min(40, len(data) // 2), ax=axes[1])  # 40 시차까지 부분자기상관 플롯
axes[1].set_title('Partial Autocorrelation Function (PACF)')  # 그래프 제목 설정

plt.tight_layout()  # 레이아웃 자동 조정
plt.show()  # 플롯 표시
```

```python
'''

RIMA 모델

ARIMA (AutoRegressive Integrated Moving Average) 모델은 시계열 데이터를 예측하기 위해 자주 사용되는 모델입니다.

ARIMA 모델의 구성 요소

ARIMA 모델은 다음 세 가지 구성 요소로 구성됩니다:

자기회귀 (AR): 이전 값들이 현재 값에 미치는 영향을 설명합니다.

차분 (I): 데이터의 비정상성을 제거하기 위해 차분을 수행합니다.

이동 평균 (MA): 오류 항의 이동 평균을 모델링합니다.



ARIMA 모델의 적합

데이터의 안정성 확인:

시계열 데이터가 안정적이어야 ARIMA 모델을 잘 적합할 수 있습니다.
데이터가 비정상적이면 차분을 통해 안정화를 수행합니다.

모델 적합:

적절한 AR, I, MA의 파라미터를 선택하여 ARIMA 모델을 적합합니다.
모델 적합 후 예측을 수행합니다.

실생활 예시:

매출 예측: 소매업체는 과거의 매출 데이터를 기반으로 미래의 매출을 예측하기 위해 ARIMA 모델을 사용할 수 있습니다.
예를 들어, 월별 판매 데이터가 있을 때, ARIMA 모델을 사용하여 다음 달의 판매량을 예측할 수 있습니다.

전력 소비 예측: 전력 회사는 과거의 전력 소비 데이터를 분석하여 미래의 전력 소비를 예측합니다.
ARIMA 모델을 사용하여 계절성과 트렌드를 고려한 예측이 가능합니다.

'''

from statsmodels.tsa.arima.model import ARIMA

# 데이터 차분 (데이터 안정화)
data['Sales_diff'] = data['Sales'].diff().dropna()

# ARIMA 모델 적합
model = ARIMA(data['Sales'], order=(1, 1, 1))  # ARIMA(p, d, q)에서 p=1, d=1, q=1 설정
fitted_model = model.fit()

# 예측 수행
forecast = fitted_model.forecast(steps=12)

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')
plt.plot(pd.date_range(start=data.index[-1] + pd.DateOffset(months=1), periods=12, freq='MS'), forecast, label='ARIMA Forecast', color='r')
plt.title('ARIMA Model Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
'''

계절적 ARIMA (SARIMA) 모델
SARIMA (Seasonal ARIMA) 모델은 계절성을 포함하는 시계열 데이터를 모델링합니다. 기본 ARIMA 모델에 계절성 구성 요소를 추가합니다.

원리 및 단계
계절성 구성 요소 추가:

SARIMA 모델은 ARIMA 모델의 계절성 버전입니다. 계절적 자기회귀(Seasonal AR), 계절적 차분(Seasonal I), 계절적 이동 평균(Seasonal MA)을 포함합니다.
모델 적합:

계절성 주기와 함께 ARIMA 모델의 파라미터를 설정하고 모델을 적합합니다.


실생활예시:

관광 수입 예측: 관광 산업에서 SARIMA 모델을 사용하여 계절적 관광 패턴을 분석하고 예측할 수 있습니다.
예를 들어, 여름 시즌 동안 관광객 수가 증가하는 패턴을 예측할 때 유용합니다.

소매업계 계절성: 소매업체는 연말 연휴와 같은 계절적 수요 패턴을 예측하기 위해 SARIMA 모델을 사용할 수 있습니다.
예를 들어, 크리스마스 시즌 동안의 판매량을 예측할 수 있습니다.

'''


from statsmodels.tsa.statespace.sarimax import SARIMAX

# SARIMA 모델 적합
sarima_model = SARIMAX(data['Sales'],
                       order=(1, 1, 1),  # ARIMA 파라미터
                       seasonal_order=(1, 1, 1, 12))  # 계절성 ARIMA 파라미터 (계절 주기 12개월)
fitted_sarima_model = sarima_model.fit()

# 예측 수행
sarima_forecast = fitted_sarima_model.forecast(steps=12)

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')
plt.plot(pd.date_range(start=data.index[-1] + pd.DateOffset(months=1), periods=12, freq='MS'), sarima_forecast, label='SARIMA Forecast', color='r')
plt.title('SARIMA Model Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
'''


지수 평활법 (Exponential Smoothing)
지수 평활법은 시계열 데이터의 미래 값을 예측하기 위해 과거 데이터를 지수적으로 가중 평균하는 방법입니다.
최근의 데이터에 더 많은 중요도를 두고, 과거 데이터는 점점 덜 중요하게 반영합니다.

단순 지수 평활법 (Simple Exponential Smoothing)

단순 지수 평활법은 시계열 데이터의 추세나 계절성을 고려하지 않고, 단순히 최근 관측값에 더 많은 가중치를 부여하여 예측합니다.
예측 값은 이전 관측값과 예측 값의 가중 평균입니다.

실생활 예시:

일일 매출 예측: 소매점에서 매일의 매출 데이터를 수집하여 향후 일일 매출을 예측할 때 사용합니다.
최근 며칠의 매출 데이터가 향후 매출 예측에 더 큰 영향을 미칩니다.



'''

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing

# 예시 데이터 생성
np.random.seed(0)
dates = pd.date_range(start='2022-01-01', end='2024-12-01', freq='MS')
n = len(dates)
data = pd.DataFrame({
    'Date': dates,
    'Sales': 100 + np.linspace(0, 50, n) + 20 * np.sin(np.linspace(0, 2 * np.pi, n)) + np.random.normal(0, 10, n)
})
data.set_index('Date', inplace=True)

# 단순 지수 평활법 모델 적합
model = ExponentialSmoothing(data['Sales'], trend=None, seasonal=None, initialization_method='estimated')
fitted_model = model.fit(smoothing_level=0.2)  # 스무딩 파라미터 설정

# 예측 수행
forecast = fitted_model.forecast(steps=12)  # 향후 12개월 예측

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')  # 원본 데이터
plt.plot(pd.date_range(start=data.index[-1] + pd.DateOffset(months=1), periods=12, freq='MS'), forecast, label='Exponential Smoothing Forecast', color='r')  # 예측 결과
plt.title('Simple Exponential Smoothing Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
'''

홀트-윈터스 모델 (Holt-Winters Model)

원리:

홀트-윈터스 모델은 단순 지수 평활법을 확장하여 계절성과 추세를 함께 고려합니다.
계절성은 주기적인 패턴을 포착하고, 추세는 장기적인 방향성을 반영합니다.

실생활 예시:

소매업계의 계절적 수요 예측: 특정 계절에 따라 판매량이 변동하는 제품의 경우,
홀트-윈터스 모델을 사용하여 계절성과 추세를 반영한 예측이 가능합니다.


'''


# 홀트-윈터스 모델 적합
holt_winters_model = ExponentialSmoothing(data['Sales'], trend='add', seasonal='add', seasonal_periods=12)
fitted_hw_model = holt_winters_model.fit()  # 모델 적합

# 예측 수행
forecast_hw = fitted_hw_model.forecast(steps=12)  # 향후 12개월 예측

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')  # 원본 데이터
plt.plot(pd.date_range(start=data.index[-1] + pd.DateOffset(months=1), periods=12, freq='MS'), forecast_hw, label='Holt-Winters Forecast', color='r')  # 예측 결과
plt.title('Holt-Winters Model Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
'''
상태공간 모델 (State Space Models)
상태공간 모델은 시계열 데이터를 상태 변수와 관측 변수로 나누어 분석하며, 관측된 데이터로부터 숨겨진 상태를 추정합니다



칼만 필터 (Kalman Filter)

원리:
칼만 필터는 데이터를 시간에 따라 추적하고 예측하는 데 사용됩니다. 항공기 위치를 추적하는 예를 들어보겠습니다.


항공기의 위치를 실시간으로 추적하고 싶습니다.
그러나 위치 측정값에는 항상 약간의 오차가 있습니다.
이 오차를 고려하여 실제 위치를 추정해야 합니다.

상태: 실제 항공기의 위치와 속도 (우리의 관심사)
관측: 측정된 위치 (오차가 포함될 수 있음)

수행단계

예측 단계: 이전 상태(항공기의 위치와 속도)를 바탕으로 현재 상태를 예측합니다.
예를 들어, 항공기가 일정 속도로 비행한다고 가정하면, 이전 위치와 속도를 바탕으로 현재 위치를 예측할 수 있습니다.

업데이트 단계: 새로운 위치 측정값이 들어오면, 예측한 위치와 실제 측정값을 비교하여 추정값을 조정합니다.
이 과정에서 예측과 관측의 신뢰도를 고려합니다.



칼만 필터는 마치 흐릿한 유리를 통해 물체를 보는 것과 같습니다.
유리가 흐릿해서 정확히 보기는 어렵지만, 유리를 통해 얻은 정보와 예측을 결합하여 물체의 위치를 보다 정확히 추정하려고 하는 방법입니다.

칼만 필터는 주어진 데이터와 모델을 사용하여 상태를 추정하는 방법입니다.
이 과정에서 예측과 관측값을 결합하여 더 정확한 추정값을 제공합니다.
항공기 위치 추적 예시에서 위치와 속도를 실시간으로 추적할 때 유용합니다.


'''

!pip install pykalman

from pykalman import KalmanFilter

# 칼만 필터 모델 적합
kf = KalmanFilter(initial_state_mean=0, n_dim_obs=1)
kf = kf.em(data['Sales'], n_iter=10)  # EM 알고리즘을 사용하여 모델 파라미터를 추정
(smoothed_state_means, smoothed_state_covariances) = kf.smooth(data['Sales'])  # 상태 추정

# 예측 수행 (향후 12개월 예측)
forecast_kf = kf.filter(data['Sales'])[0]  # 필터를 사용하여 예측

# 시각화
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')  # 원본 데이터
plt.plot(data.index, forecast_kf, label='Kalman Filter Forecast', color='r')  # 예측 결과
plt.title('Kalman Filter Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
'''

베이지안 상태공간 모델 (Bayesian State Space Models)


베이지안 상태공간 모델은 데이터의 불확실성을 다루는 방법입니다.
베이지안 방법은 사전 정보와 새로운 데이터를 결합하여 예측을 개선합니다. 여기서는 매출 예측을 예로 들어 설명하겠습니다.

매월 매출을 예측하려고 합니다. 하지만 매출 예측에는 불확실성이 있습니다.
예를 들어, 새로운 프로모션이 있을 때와 없을 때 매출이 달라질 수 있습니다.

사전 분포와 사후 분포:

사전 분포: 우리가 예측을 시작할 때 가지고 있는 기본적인 정보입니다.
예를 들어, 과거의 평균 매출에 대한 정보가 있을 수 있습니다.

사후 분포: 새로운 데이터를 수집한 후 업데이트된 정보입니다.
새로운 매출 데이터가 들어오면, 사전 정보와 결합하여 예측을 조정합니다.


모델 정의: 매출 데이터의 변동을 설명할 수 있는 모델을 정의합니다. 예를 들어, 매출은 계절적 요인과 프로모션에 따라 변할 수 있습니다.

샘플링: 모델에 의해 추정된 다양한 가능성을 샘플링합니다. 이렇게 하면 매출 예측에 대한 다양한 시나리오를 고려할 수 있습니다.

예측: 새로운 데이터가 들어올 때마다 예측을 업데이트합니다. 이 과정에서 데이터의 불확실성을 반영합니다.


베이지안 상태공간 모델은 마치 날씨 예보와 같습니다.
날씨 예보는 과거의 기상 데이터와 현재의 날씨 상황을 바탕으로 미래의 날씨를 예측합니다.
예보가 업데이트되면 새로운 데이터가 반영되어 예측이 조정됩니다.
불확실성을 고려하여 다양한 가능성을 제공합니다.

베이지안 상태공간 모델은 데이터의 불확실성을 고려하여 예측을 수행합니다.
사전 정보와 새로운 데이터를 결합하여 예측을 업데이트합니다.
매출 예측과 같은 상황에서 불확실성을 효과적으로 다룰 수 있습니다.


'''
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import cholesky, solve_triangular

# 예시 데이터 생성
np.random.seed(0)  # 난수 생성의 시드를 설정하여 결과의 재현성을 보장합니다.
dates = pd.date_range(start='2022-01-01', end='2024-12-01', freq='MS')  # 월 단위의 날짜 범위를 생성합니다.
n = len(dates)  # 날짜의 개수를 저장합니다.
data = pd.DataFrame({
    'Date': dates,  # 생성된 날짜를 'Date' 열로 추가합니다.
    'Sales': 100 + np.linspace(0, 50, n) + 20 * np.sin(np.linspace(0, 2 * np.pi, n)) + np.random.normal(0, 10, n)  # 매출 데이터 생성
    # 기본 매출 + 선형 증가 + 계절성 (사인 함수) + 잡음
})
data.set_index('Date', inplace=True)  # 'Date' 열을 인덱스로 설정합니다.

# 상태공간 모델 정의
def bayesian_state_space(y, F, H, Q, R):
    """
    간단한 베이지안 상태공간 모델을 구현하는 함수입니다.

    :param y: 관측값 (시계열 데이터)
    :param F: 상태 전이 행렬
    :param H: 관측 행렬
    :param Q: 과정 잡음 공분산
    :param R: 관측 잡음 공분산
    :return: 상태 추정값과 공분산
    """
    n = len(y)  # 시계열 데이터의 길이
    m = F.shape[0]  # 상태 변수의 차원

    # 초기 상태와 공분산 설정
    x = np.zeros((m, n))  # 상태 변수 저장 배열 (초기 상태는 0으로 설정)
    P = np.zeros((m, m, n))  # 공분산 저장 배열
    x[:, 0] = np.zeros(m)  # 초기 상태를 0으로 설정
    P[:, :, 0] = np.eye(m)  # 초기 공분산을 단위 행렬로 설정

    # 베이지안 추정 초기화
    mu = np.zeros(m)  # 상태 추정값 초기화
    Sigma = np.eye(m)  # 공분산 초기화

    # 필터링 과정
    for k in range(1, n):
        # 예측 단계
        mu = F @ mu  # 상태 예측
        Sigma = F @ Sigma @ F.T + Q  # 공분산 예측

        # 업데이트 단계
        K = Sigma @ H.T @ np.linalg.inv(H @ Sigma @ H.T + R)  # 칼만 이득 계산
        mu = mu + K @ (y[k] - H @ mu)  # 상태 업데이트
        Sigma = Sigma - K @ H @ Sigma  # 공분산 업데이트

        x[:, k] = mu  # 상태 추정값 저장
        P[:, :, k] = Sigma  # 공분산 저장

    return x, P  # 상태 추정값과 공분산 반환

# 상태공간 모델 파라미터 설정
F = np.array([[1, 1], [0, 1]])  # 상태 전이 행렬: 현재 상태와 이전 상태의 관계를 정의합니다.
H = np.array([[1, 0]])  # 관측 행렬: 상태에서 관측값을 추출하는 방식을 정의합니다.
Q = np.array([[1, 0], [0, 1]])  # 과정 잡음 공분산: 상태 전이 과정의 잡음을 나타냅니다.
R = np.array([[10]])  # 관측 잡음 공분산: 관측 과정에서의 잡음을 나타냅니다.

# 상태공간 모델 적용
x, P = bayesian_state_space(data['Sales'].values, F, H, Q, R)  # 상태공간 모델을 데이터에 적용하여 상태 추정값과 공분산을 계산합니다.

# 예측 수행
forecast_dates = pd.date_range(start=data.index[-1] + pd.DateOffset(months=1), periods=12, freq='MS')  # 예측을 위한 날짜 범위를 생성합니다.
forecast_mean = x[0, -1] + np.linspace(0, 20, 12)  # 예측값: 필터링된 마지막 상태에 선형 증가를 추가

# 시각화
plt.figure(figsize=(12, 6))  # 그래프의 크기를 설정합니다.
plt.plot(data.index, data['Sales'], label='Original Sales Data', color='b')  # 원본 데이터 시각화 (파란색)
plt.plot(data.index, x[0, :], label='Filtered Sales Data', color='g', linestyle='--')  # 필터링된 데이터 시각화 (녹색, 점선)
plt.plot(forecast_dates, forecast_mean, label='Forecast Sales Data', color='r', linestyle='--')  # 예측 결과 시각화 (빨간색, 점선)
plt.title('Bayesian State Space Model Forecast')  # 그래프 제목 설정
plt.xlabel('Date')  # x축 레이블 설정
plt.ylabel('Sales')  # y축 레이블 설정
plt.legend()  # 범례 표시
plt.show()  # 그래프 출력
```
